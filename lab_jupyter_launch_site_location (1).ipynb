Next, let's try to enhance the map by adding the launch outcomes for each site, and see which sites have high success rates.
Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
import folium

# Create base map centered near Cape Canaveral
site_map = folium.Map(location=[28.563, -80.576], zoom_start=5)

# Group by LaunchSite to calculate success rate
site_success = spacex_df.groupby("Launch Site")["class"].mean().reset_index()
site_counts = spacex_df.groupby("Launch Site")["class"].count().reset_index()

# Merge counts and success rates
site_summary = site_success.merge(site_counts, on="Launch Site")
site_summary.columns = ["Launch Site", "SuccessRate", "TotalLaunches"]

# Coordinates for each site (from earlier task)
launch_sites = {
    "CCAFS LC-40": [28.562302, -80.577356],
    "CCAFS SLC-40": [28.563197, -80.576820],
    "KSC LC-39A": [28.573255, -80.646895],
    "VAFB SLC-4E": [34.632834, -120.610745]
}

# Add circles with success rate info
for _, row in site_summary.iterrows():
    site = row["Launch Site"]
    success_rate = row["SuccessRate"]
    total_launches = row["TotalLaunches"]
    coord = launch_sites[site]
    
    # Color intensity based on success rate
    color = "green" if success_rate > 0.7 else "orange" if success_rate > 0.4 else "red"
    
    folium.Circle(
        location=coord,
        radius=2500,
        color=color,
        fill=True,
        fill_color=color,
        popup=f"{site}\nSuccess Rate: {success_rate:.2f}\nTotal Launches: {total_launches}"
    ).add_to(site_map)

# Display map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from geopy.distance import geodesic
launch_sites = {
    "CCAFS LC-40": [28.562302, -80.577356],
    "CCAFS SLC-40": [28.563197, -80.576820],
    "KSC LC-39A": [28.573255, -80.646895],
    "VAFB SLC-4E": [34.632834, -120.610745]
}
proximities = {
    "Orlando Airport": [28.4312, -81.3081],
    "Los Angeles Airport": [33.9416, -118.4085],
    "Cape Canaveral": [28.3922, -80.6077]
}
distances =[]
for site, site_coord in launch_sites.items():
    for prox, prox_coord in proximities.items():
        dist_km = geodesic(site_coord, prox_coord).km
        distances.append([site, prox, dist_km])
distance_df = pd.DataFrame(distances, columns=["Launch Site","Proximity","Distance_km"])
distance_df.head()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
closest_city = (28.38415, -80.60313)
closest_highway = (28.56222, -80.57066)
closest_railway = (28.56275, -80.58704)
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=5)
```

```python
# Add marker for launch site
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site: CCAFS LC-40",
    icon=folium.Icon(color="blue", icon="rocket")
).add_to(site_map)

# Function to add proximity marker + distance line
def add_proximity_marker(site_coord, prox_coord, label):
    distance = calculate_distance(site_coord, prox_coord)
    
    # Marker for proximity point
    folium.Marker(
        prox_coord,
        popup=f"{label}\nDistance: {distance:.2f} km",
        icon=folium.Icon(color="green", icon="info-sign")
    ).add_to(site_map)
    
    # Distance label marker
    folium.Marker(
        prox_coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
        )
    ).add_to(site_map)
    
    # Line between site and proximity point
    folium.PolyLine(
        locations=[site_coord, prox_coord],
        color="red",
        weight=2,
        dash_array="5, 5"
    ).add_to(site_map)

# Add city, highway, railway markers
add_proximity_marker((launch_site_lat, launch_site_lon), closest_city, "Closest City")
add_proximity_marker((launch_site_lat, launch_site_lon), closest_highway, "Closest Highway")
add_proximity_marker((launch_site_lat, launch_site_lon), closest_railway, "Closest Railway")

# Display map
site_map
```
